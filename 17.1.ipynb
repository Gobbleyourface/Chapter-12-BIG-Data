```python
# connects with the databse
import sqlite3
import pandas as pd
connection = sqlite3.connect('books.db')
```

```python
# Displays Author's Last names in descending order
pd.read_sql('SELECT last FROM authors ORDER BY last DESC', connection)
```

```python
# Displays all book titles in ascending oder
pd.read_sql('SELECT title FROM titles',connection)
```

```python
# Displays the title, copyright year, and ISBN based on a specific author
pd.read_sql("""SELECT title, copyright, titles.isbn
               FROM titles
               INNER JOIN author_ISBN ON titles.isbn = author_ISBN.isbn
               WHERE author_ISBN.id = 1
               ORDER BY title""", connection)
```

```python
# Insert a new author into the authors table
cursor = connection.cursor()
cursor = cursor.execute("""INSERT INTO authors (first, last) VALUES ('Jiawei', 'Chen')""")
pd.read_sql('SELECT id, first, last FROM authors', connection, index_col = ['id'])
```
